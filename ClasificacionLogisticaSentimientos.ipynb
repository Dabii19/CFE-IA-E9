```python
#SE IMPORTAN LOS DATOS Y SE OBSERVAN LOS PRIMEROS 5

import pandas as pd

df = pd.read_csv("dataset_train.csv")

df.head(5)
```

```python
# SE MUESTRAN LA CANTIDAD DE DATOS NO VACIOS

df.info()
```

```python
#MOSTRAMOS LOS DATOS NULOS SI LOS HAY

df.isnull().sum()
```

```python
#ANALIZAMOS LOS DUPLICADOS Y ELIMINAMOS

print ("Cantidad de filas antes de eliminar los duplicados", df.shape)

df.drop_duplicates(inplace = True)

print ("Cantidad de filas después de eliminar", df.shape)
```

```python
#SE CREA EL MODELO

from sklearn.model_selection import train_test_split

#VARIABLES A USAR (TEXT AND EMOTIONS, ADEMAS SE USA UN 10% PARA ENTRENAR Y EL RESTO PARA PROBAR)

X_train, X_test, y_train, y_test = train_test_split(df['text'], df['emotion'], test_size = 0.10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizador = CountVectorizer()

X_train_vec = vectorizador.fit_transform(X_train)

X_test_vec = vectorizador.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

modelo = LogisticRegression()

modelo.fit (X_train_vec, y_train)

predict = modelo.predict (X_test_vec)

accuaracy = modelo.score (X_test_vec, y_test)

print ("Porcentaje de éxito", accuaracy * 100, " %")
```

```python
#CARGA Y MUESTRA DEL ARCHIVO DE PRUEBA

test_data = pd.read_csv("dataset_test_sin_sentimento.csv")

test_data.head(5)
```

```python
X_test_vec = vectorizador.transform(test_data['text'])

predict = modelo.predict(X_test_vec)

output = pd.DataFrame({'ID': test_data['id'], 'sentimiento': predict})

output.to_csv('dataset_test_prueba.csv', index = False)

print ("File created successfully")
```
